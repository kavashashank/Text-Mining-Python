### 1. Create a bar chart of the number of restaurant inspections in each of the five NYC boroughs. Be sure to include a chart title, make the bars blue, and set the font size of the labels to 14pt.

```python
#Load the csv file in a dataframe
rest_inspect_df=pd.read_csv("Datasets/New_York_City_Restaurant_Inspections.csv")
```

```python
#Plots bar chart for number of restaurants in each borough with aesthetics
rest_inspect_df.BORO.value_counts().plot(kind="bar",color="blue", title="#Restaurants in NYC Boroughs", fontsize=14)
```

### 2. Create a histogram of the restuarant scores to show the distribution of values. 
Set the number of bins to 20 and make sure that the range of the x-axis starts at zero.

```python
#Plots histograms for restaurant score with bins as an attribute
axes=rest_inspect_df.hist(column="SCORE", bins=20, figsize=(16,4))
axes[0][0].set_xlabel("Score for a Restaurant")
```

### 3. Create a set of histograms to compare the distribution of scores across the five boroughs.

```python
#Plots seperate histograms for the score for each borough
rest_inspect_df.hist(column="SCORE", by="BORO", figsize=(16,20), sharey=True)
```

### 4. Plot a chart showing the number of inspections per day of the week broken out by borough (i.e. one line per borough). 
Be sure to include a legend indicating a label for each colored line.

```python
from datetime import datetime

type(rest_inspect_df["INSPECTION DATE"][0])

#Returns the day for each inspection date in the datafrae
def get_day(x):
    x = datetime.strptime(x["INSPECTION DATE"], '%m/%d/%Y').strftime('%A')
    return x

rest_inspect_df["DAY"] = rest_inspect_df.apply(get_day, axis=1)
```

```python
#Plots line graph for number of inspections per day for each borough
rest_inspect_df[rest_inspect_df["BORO"] == "BRONX"].DAY.value_counts().plot(figsize=(15,5))
rest_inspect_df[rest_inspect_df["BORO"] == "BROOKLYN"].DAY.value_counts().plot(figsize=(15,5))
rest_inspect_df[rest_inspect_df["BORO"] == "MANHATTAN"].DAY.value_counts().plot(figsize=(15,5))
rest_inspect_df[rest_inspect_df["BORO"] == "QUEENS"].DAY.value_counts().plot(figsize=(15,5))
rest_inspect_df[rest_inspect_df["BORO"] == "STATEN ISLAND"].DAY.value_counts().plot()
plt.legend(["BRONX","BROOKLYN","MANHATTAN","QUEENS","STATEN ISLAND"], loc="best")
```
