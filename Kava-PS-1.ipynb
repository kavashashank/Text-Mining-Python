### 1. Get Data and Load it
Download the [datafile](https://github.com/comp-journalism/UMD-J479V-J779V-Spring2017/blob/master/Data/911_Calls_Baltimore_11-2016.csv?raw=true) of the 911 calls in the City of Baltimore from November, 2016. 

Load the .csv file into a variable named "`calls_df`" and print it out. There is a column called `callDateTime` that you should be sure to parse as a datetime field. 

```python
import os #Imports the OS library for functionality related to the operating system
os.chdir('C:\Users\kavas\Desktop\MyOwnDocs\Courses\Sem2\JOUR779V\Datasets') #Changes the current working directory

import pandas as pd #Imports pandas library

#loads the file and parses column callDateTime as a datetime field
calls_df = pd.read_csv("911_Calls_Baltimore_11-2016.csv",parse_dates=["callDateTime"])
calls_df
```

### 2. Remove Columns and Save as a new .csv file
Remove the "callNumber" and "location" columns from the `calls_df` dataframe. Then save the remaining columns to a new .csv file being sure to exclude the index column from output as well. 

```python
#drop function to remove columns specified as parameter
calls_df.drop(['callNumber','location'],axis=1,inplace=True)
#axis = 1 indicates column
#inplace to delete without reassigning 

#ouputs a dataframe to .csv file in the current working directory
calls_df.to_csv('Test_outputfile.csv',index=False)
```

### 3. What are the different unique values that the "priority" column can take on? What about for the "district" column?
If you look in the `priority` column you see different values like "low", "medium" etc. What is the complete set of different values that you find in that column, or in the `district` column?

```python
#prints the complete set of different values for priority column
priority_unique = calls_df.priority.unique()
print priority_unique

#prints the complete set of different values for district column
district_unique = calls_df.district.unique()
print district_unique
```

### 4. Filter the dataframe to only calls that were "Low" or "Medium" priority. How many were there? What fraction of the total do these represent?

```python
#Filtering records using the 'OR' condition
calls_df_filtered=calls_df[(calls_df.priority=="Low")|(calls_df.priority=="Medium")]

#Returns the number of rows for the filtered dataset
count_filtered=calls_df_filtered.shape[0]
print "There are",count_filtered,"records in the filtered dataset."

#Returns the number of rows for the original dataset
count_calls_df=calls_df.shape[0]

#Returns what fraction of the total does the filtered set represent
fraction_filtered= count_filtered/float(count_calls_df)
print "It represents",fraction_filtered,"of the original dataframe"
```

### 5. Group the dataframe by district and summarize the total number of calls by neighborhood. Which neighborhood district gets the most number of calls?

```python
#Groups the dataframe based on values in the district column
grouped_calls_df = calls_df.groupby("district").size()

#Size function summarizes the total number or calls by neighborhood
print grouped_calls_df

#Prints the neighborhood that gets most calls
print "Neighborhood district that gets the most number of calls is",grouped_calls_df.argmax(),":",grouped_calls_df.max()



```
